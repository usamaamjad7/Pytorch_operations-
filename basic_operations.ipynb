# Tensor Broad Casting
the rules of broadcasting are....

*   Each tensor must have atleast one dimension .. no empty tensors
*   Comparing the dimension sizes of two tensor going from last to first.

*   Each dimension must be equal ,or
*   One of the dimension must be size of 1

*   The dimension doesnot exist in one of the tensors

```python
import math
```

```python

# common functions
a = torch.rand(2,4)*2 -1
print("Common Functions")
print(torch.abs(a))
print(torch.ceil(a))
print(torch.floor(a))
print(torch.clamp(a,-0.5,0.5))

# trignometric functions and their inverses
angles = torch.tensor([0, math.pi / 4,math.pi/2,3 * math.pi/4])
sines = torch.sin(angles)
inverses = torch.asin(sines)
print("\nSine and arcsine")
print(angles)
print(sines)
print(inverses)

#biwise operation..
print("\n Bitwise XOR")
b = torch.tensor([1,5,11])
c = torch.tensor([2,7,10])
print(torch.bitwise_xor(b,c))

#Comparisons..
print("\n Broadcasted-element wise equality Comparison")
d = torch.tensor([[1.,2.],[3.,4.]])
e = torch.ones(1,2)
print(torch.eq(d,e))

# reductions..
print("\n Broadcasted , element wise equality Comparison")
print(torch.max(d))
print(torch.max(d).item())
print(torch.mean(d))
print(torch.std(d))
print(torch.prod(d))
print(torch.unique(torch.tensor([1,2,1,2,1,2])))

# vector and Linear algebra operations..
print("\n Reduction ops")
v1 = torch.tensor([1.,0.,0.])
v2 = torch.tensor([0.,1.,0.])
m1 = torch.rand(2,2)
m2 = torch.tensor([[3.,0.],[0.,3.]])

print("\n vector & matrices")
print(torch.cross(v2,v1))
print(m1)
m3 = torch.matmul(m1,m2)
print(m3)
print(torch.svd(m3))
```

# Underscore _()

```python
a = torch.tensor([0, math.pi / 4,math.pi/2,3 * math.pi/4])
print("a")
print(a)
print(torch.sin(a)) # creates new tensor, but a has not changed
print(a)

print("b")
b = torch.tensor([0, math.pi / 4,math.pi/2,3 * math.pi/4])
print(b)
print(torch.sin_(a))
print(a)
```

# Functions that behave similarly

```python
a = torch.rand(2,2)
b = torch.rand(2,2)

print("Before")
print(a)
print(b)

print("\n After adding")
print(a.add_(b))
print(a)
print(b)

print("after  multiplying")
print(b.mul_(b))
print(b)
```

# Another..

```python
a = torch.rand(2,2)
b = torch.rand(2,2)
c = torch.rand(2,2)
old_id = id(c) # get the id of the object..

print(c)
d = torch.matmul(a,b ,out=c)
print(c)

assert c is d
assert id(c),old_id

torch.rand(2,2, out=c)
print(c)
assert id(c),old_id
```

# The clone method

```python
a = torch.ones(2,2)
b = a.clone()

assert b is not a
print(torch.eq(a,b))
a[0][1] = 561
print(b)
```

# Aware of when using clone ()

```python
a = torch.ones(2,2, requires_grad=True)
print(a)
b =  a.clone()
print(b)
c = a.detach().clone()
print(c)

print(a)
```
